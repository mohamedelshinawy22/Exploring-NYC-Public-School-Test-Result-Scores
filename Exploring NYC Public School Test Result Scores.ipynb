Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# 1. Find schools with the best math results (at least 80% of 800)
# Assuming the column 'average_math' contains the average math score
threshold = 0.8 * 800
best_math_schools = schools[schools['average_math'] >= threshold][['school_name', 'average_math']].sort_values(by='average_math', ascending=False)

# 2. Find the top 10 performing schools based on the combined SAT scores
# Assuming the columns 'average_math', 'average_reading', and 'average_writing' contain the SAT scores
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
top_10_schools = schools[['school_name', 'total_SAT']].sort_values(by='total_SAT', ascending=False).head(10)

# 3. Find the borough with the largest standard deviation in combined SAT scores
# Group by borough and calculate statistics
borough_stats = schools.groupby('borough')['total_SAT'].agg(['mean', 'std', 'count']).reset_index()
borough_stats.columns = ['borough', 'average_SAT', 'std_SAT', 'num_schools']

# Find the borough with the maximum standard deviation
largest_std_dev = borough_stats.loc[borough_stats['std_SAT'].idxmax()][['borough', 'num_schools', 'average_SAT', 'std_SAT']]
largest_std_dev = pd.DataFrame([largest_std_dev])

# Round all numeric values to two decimal places
best_math_schools['average_math'] = best_math_schools['average_math'].round(2)
top_10_schools['total_SAT'] = top_10_schools['total_SAT'].round(2)
largest_std_dev[['average_SAT', 'std_SAT']] = largest_std_dev[['average_SAT', 'std_SAT']].round(2)

# Display the results
print("Best Math Schools:")
print(best_math_schools)

print("\nTop 10 Performing Schools Based on Combined SAT Scores:")
print(top_10_schools)

print("\nBorough with the Largest Standard Deviation in Combined SAT Scores:")
print(largest_std_dev)
```
